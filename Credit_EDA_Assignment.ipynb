**Step-2** - Standardising & Binning Values



```python
app_data.head()
```

We now see that the number of days are negative value in columns "DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH", "DAYS_LAST_PHONE_CHANGE". 

We will convert these values into postitive values and to years by diving the days.

```python
#Coverting the negative day values to absolute values
app_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH"]] = app_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH"]].abs()
```

```python
#Coverting the no of days to years and round upto 1 place decimal
app_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH"]] = round(app_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH"]]/365, 1)
```

```python
#Renaming columns names "DAYS" to "YEARS" 
app_data.rename(columns = {"DAYS_BIRTH":"YEARS_BIRTH", "DAYS_EMPLOYED":"YEARS_EMPLOYED", "DAYS_REGISTRATION": "YEARS_REGISTRATION", "DAYS_ID_PUBLISH":"YEARS_ID_PUBLISH"}, inplace = True)
```

```python
#Changing "DAYS_LAST_PHONE_CHANGE" to absolute values from negative values (if any) and converting to years
app_data["DAYS_LAST_PHONE_CHANGE"] = round(abs(app_data["DAYS_LAST_PHONE_CHANGE"]/365), 1)
```

```python
app_data.rename(columns = {"DAYS_LAST_PHONE_CHANGE":"YEARS_LAST_PHONE_CHANGE"}, inplace = True)
```
